La funci贸n filter() filtra una lista de elementos para los cuales una funci贸n devuelve Verdadero.
```python
filter(a_function, a_list)
```
Nos puede dar la funci贸n:
- `True`
- `Truthy`: obj no booleano pero la evaluaci贸n es verdadera.

#### Primero lo hacemos con funciones

Imagina que quieres filtrar una lista de n煤meros para obtener solo valores pares.

````python
def pares_impares (n):
    # Filter: devuelve true o false
    if n%2 == 0: 
        return True
    return False
````

````python
lst_num = [2, 324, 45, 24, 456, 32, 100, 99, 88, 7, 1, 1, 1, 1, 3, 2]
list(filter(pares_impares, lst_num))
````

````python
def pares_impares (n):
    # Filter: devuelve true o false
    if n%2 == 0: #True
        return n**2 #truthy
    return False
print(list(map(pares_impares, lst_num)))
list(filter(pares_impares, lst_num)) #Te devuelve los que evalua en verdadero
````

### Reduce
Itera y reduce

````python
from functools import reduce
````

Esta funci贸n no es una funci贸n incorporada, lo que significa que primero debe importar la biblioteca functools para usarla. La funci贸n ``
`reduce()` aplica de forma acumulativa la funci贸n f() a los elementos del iterable. El acumulador se puede inicializar con un tercer argumento opcional. El resultado es el valor final del acumulador.

```python
reduce(function, iterable[, initial])
```

````python
def suma(a, b):
    return a + b
````

````python
lst_num = [1, 3, 6, 2, 5, 67]
'''
lst_num = [1, 3, 6, 2, 5, 67]
lst_num = [10, 2, 5, 67]
lst_num = [12, 5, 67]
lst_num = [17, 67]
lst_num = 84
'''
reduce(suma,lst_num)
````

````python
reduce(suma, ciudades)
````

````python
reduce(lambda a, b: a + "--" + b, ciudades)
````

#### 驴Qu茅 sucedi贸?
- Empezamos con una lista [2, 4, 7, 3] y pasamos la funci贸n add (x, y) a reduce ( ) junto con esta lista, sin un valor inicial

- reduce() llama a add(2, 4), y add() devuelve 6

- reduce() llama a sum(6, 7) (resultado de la llamada anterior a sum() y el siguiente elemento en la lista como par谩metros), y sum() devuelve 13

- reduce() llama a sum(13, 3), y sum() devuelve 16

Como no quedan m谩s elementos en la secuencia, reduce() devuelve 16
