## 7. Write a function that takes a list and returns a list of the unique values.

Forbidden built-in functions: `sum()`. You can use `lst.append`

```python
def unique(arr):
    nueva = []
    for pos, i in enumerate(arr[:]):  # recorremos una copia con arr[:]
        if i not in nueva:
            nueva.append(i)
        for pos2, j in enumerate(arr[pos+1:], start=pos+1):
            if i == j:
                arr.pop(pos2)
                break  # salir para evitar error de √≠ndice
    return nueva
```

## 11. Write a function that returns the mode of a list, i.e.: the element that appears the most times.
`NOTE: You should not use count... üßê`

```python
def mode_counter(arr):
    arg_set = set(arr)
    dicc = {}
    for i in arg_set:
        dicc[i]= arr.count(i)
    max_count = max(dicc.values())
    for caract, cont in dicc.items():
        if cont == max_count:
            print(f"el simbolo {caract} es el que m√°s aparece, apareciendo este {cont}")

# Lista de prueba 1
lista1 = [1, 2, 2, 3, 4, 2, 5]
mode_counter(lista1)  # ‚Üí Deber√≠a devolver 2

# Lista de prueba 2
lista2 = ["manzana", "pera", "manzana", "banana", "pera", "manzana"]
mode_counter(lista2)  # ‚Üí Deber√≠a devolver "manzana"

# Lista de prueba 3 (con empate)
lista3 = [7, 7, 8, 8, 9]
mode_counter(lista3)  # ‚Üí Puede devolver 7 o 8, cualquiera de los dos est√° bien
```

## 12. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! ü§™`

```python
def sort_alpha(string):
    palabras = []
    palabra_actual = ""
    for char in string:
        if char == ",":
            palabras.append(palabra_actual)  # cuando hay coma, guardamos la palabra
            palabra_actual = ""  # reiniciamos
        else:
            palabra_actual += char
    palabras.append(palabra_actual)  # agregar la √∫ltima palabra

    # ordenar usando sorted
    palabras_ordenadas = sorted(palabras)

    # imprimir de forma manual con comas, sin usar join
    for i, palabra in enumerate(palabras_ordenadas):
        if i != len(palabras_ordenadas)-1:
            print(palabra, end=",")
        else:
            print(palabra)  # la √∫ltima palabra sin coma
```

#### Crear anagrama 
dos palabras son anagramas si al ordenar sus letras son iguales, es decir estan compuestas por los mismas simbolos indistintamente si son mayusculas o minusculas.

```python
# write the function is_anagram
def is_anagram(test, original):
    if test.lower()==original.lower():
        return False
    return sorted(test.lower()) == sorted(original.lower())
```
