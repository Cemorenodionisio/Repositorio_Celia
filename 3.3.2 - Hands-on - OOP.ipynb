# Ejercicio 2: Gesti√≥n de Pedidos

Crea una clase Producto y una clase Pedido que permita agregar productos y calcular el total.

```python
class Producto:
    def __init__(self, disponibilidad, productos):
        self.disponibilidad = disponibilidad              # lista de cantidades en stock
        self.productos = list(productos.keys())            # nombres de productos
        self.costes = list(productos.values())             # precios de productos

class Pedido(Producto):
    def __init__(self, disponibilidad, productos):
        super().__init__(disponibilidad, productos)

    def compra(self, pedido):
        importe = 0
        for nombre, cantidad in pedido.items():
            restante = 0
            if nombre in self.productos:
                indice = self.productos.index(nombre)          # buscar la posici√≥n del producto
                if cantidad <= self.disponibilidad[indice]:    # verificar que hay suficiente stock
                    importe += cantidad * self.costes[indice]
                elif cantidad > self.disponibilidad[indice]:
                    restante =cantidad - self.disponibilidad[indice]
                    importe += self.disponibilidad[indice] * self.costes[indice]
                    print(f"Solo has podido comprar {self.disponibilidad[indice]} {nombre}s, por tanto hay {restante} {nombre}s pendientes de reposici√≥n")
                    
        return importe
```

```python
# Datos
lst_prod = {"Cama": 1200, "Sofa": 2400, "Lampara": 20}
inventario = [2, 3, 1]

mipedido = {"Cama": 3, "Lampara": 4, "Sofa": 2}

# Crear objetos
Productos = Producto(inventario, lst_prod)
Pedidos = Pedido(inventario, lst_prod)

# Calcular total
total = Pedidos.compra(mipedido)
print("Importe total del pedido:", total)
```

Crea una clase Libro con t√≠tulo, autor y disponibilidad. Implementa una clase Biblioteca que gestione pr√©stamos y devoluciones.


```python
class Libro ():
    def __init__(self, titulo, autor):
        self.titulo = titulo
        self.autor = autor
        self.disponibilidad = True
        
class Biblioteca (Libro):
    def __init__(self, titulo, autor):
        super().__init__(titulo, autor)
    def prestamo (self, acci√≥n):
        if acci√≥n == "devoluci√≥n" and  self.disponibilidad == False:
            self.disponibilidad = True
            return f"Gracias por devolver el libro {self.titulo}"
        elif acci√≥n =="pr√©stamo" and self.disponibilidad == True :
            self.disponibilidad = False
            return f"Tienes 30 d√≠as para devolver el libro {self.titulo}"
        elif acci√≥n =="pr√©stamo" and self.disponibilidad == False :
            return f"El libro {self.titulo}, no esta disponible"
        elif acci√≥n =="devoluci√≥n" and self.disponibilidad == True :
            return f"El libro {self.titulo} no lo puedes tener, ya que esta disponible"
```

# Ejercicio 4: Sistema de Reservas de Hotel

Crea una clase Habitaci√≥n con n√∫mero, tipo (individual/doble) y disponibilidad. Luego, crea una clase Hotel que permita reservar y liberar habitaciones.

```python
class Habitacion:
    def __init__(self, numero, tipo, disponible=True):
        self.numero = numero
        self.tipo = tipo
        self.disponible = disponible

    def __str__(self):
        estado = "Disponible" if self.disponible == True else "Ocupada"
        return f"Habitaci√≥n {self.numero} ({self.tipo}) - {estado}"

class Hotel ():
    def __init__(self):
        self.habitaciones = []
    def agregar_habitacion(self, habitacion):
        self.habitaciones.append(habitacion)
    def reservar(self, numero):
        for hab in self.habitaciones:
            if hab.numero == numero:
                if hab.disponible == True:
                    hab.disponible = False
                    print(f"Habitaci√≥n {numero} reservada con √©xito.")
                else:
                    print(f"Habitaci√≥n {numero} ya est√° ocupada.")
                return
        print(f"Habitaci√≥n {numero} no encontrada.")

    def liberar(self, numero):
        for hab in self.habitaciones:
            if hab.numero == numero:
                if not hab.disponible:
                    hab.disponible = True
                    print(f"Habitaci√≥n {numero} liberada.")
                else:
                    print(f"Habitaci√≥n {numero} ya estaba libre.")
                return
        print(f"Habitaci√≥n {numero} no encontrada.")

    def mostrar_habitaciones(self):
        for hab in self.habitaciones:
            print(hab)
```

```python
# Ejemplo de uso
hotel = Hotel()
hotel.agregar_habitacion(Habitacion(101, "individual"))
hotel.agregar_habitacion(Habitacion(102, "doble"))
hotel.agregar_habitacion(Habitacion(103, "doble", disponible=False))

#hotel.mostrar_habitaciones()
```

```python
hotel.mostrar_habitaciones()
hotel.reservar(102)
hotel.liberar(103)
```

# Ejercicio 5: Gesti√≥n de Cuentas Bancarias

Implementa una clase CuentaBancaria con saldo y m√©todos para depositar, retirar y transferir dinero.

```python
class Registro ():
    def __init__(self, propietario, n_cuenta, saldo):
        self.propietario = propietario
        self.n_cuenta = n_cuenta
        self.saldo = saldo

class CuentaBancaria ():
    def __init__(self):
        self.clientes =[]
    def agregar_cuentas (self, registro):
        self.clientes.append(registro)

    def accion (self, n_cuenta, accion, importe):
        for clt in self.clientes:
            if n_cuenta == clt.n_cuenta:
                if accion == "depositar":
                    clt.saldo += importe
                elif accion == "retirar":
                    clt.saldo -= importe
                elif accion == "transferir":
                    clt.saldo -= importe
                    prop_transferencia = input("A quien se lo quieres transferir")
                    for receptor in self.clientes:
                        if receptor.propietario == prop_transferencia:
                            receptor.saldo += importe
                            print(f"Transferencia realizada a {receptor.propietario}. En tu cuenta queda {clt.saldo}")
                            break
                    else:
                        # Este else se ejecuta solo si el bucle no encontr√≥ a nadie
                        print("No se encontr√≥ al destinatario.")
                return f"En tu cuenta queda: {clt.saldo}"
    def mostrar_clientes (self):
        for i in self.clientes:
            print(f"En la cuenta {i.n_cuenta}, correspondiente a {i.propietario} quedan {i.saldo} euros")
```

# Piedra, Papel o Tijera

```python
class PiedraPapelTijera:
    def __init__(self):
        self.resultado_jugador1 = 0
        self.resultado_jugador2 = 0

    def juego(self):
        cont = 0
        while cont < 4:
            eleccion1 = input("Elecci√≥n jugador 1 (Piedra, Papel, Tijera): ").capitalize()
            eleccion2 = input("Elecci√≥n jugador 2 (Piedra, Papel, Tijera): ").capitalize()

            if eleccion1 == eleccion2:
                print("Empate.")
            elif (eleccion1 == "Piedra" and eleccion2 == "Tijera") or \
                 (eleccion1 == "Papel" and eleccion2 == "Piedra") or \
                 (eleccion1 == "Tijera" and eleccion2 == "Papel"):
                print("Gana jugador 1.")
                self.resultado_jugador1 += 1
            else:
                print("Gana jugador 2.")
                self.resultado_jugador2 += 1

            cont += 1

        if self.resultado_jugador1 > self.resultado_jugador2:
            print("üèÜ Ha ganado el jugador 1!")
        elif self.resultado_jugador2 > self.resultado_jugador1:
            print("üèÜ Ha ganado el jugador 2!")
        else:
            print("ü§ù Empate final.")

# Crear juego y ejecutar
juego = PiedraPapelTijera()
juego.juego()
```
