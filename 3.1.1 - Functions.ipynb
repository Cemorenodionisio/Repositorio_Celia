### `lambda` como opci√≥n de algunas otras funciones integradas

Tambi√©n es posible utilizar estas funciones an√≥nimas como argumentos en llamadas a otras funciones. Por ejemplo, en una llamada a sorted() o list.sort() podemos utilizar una funci√≥n lambda como argumento del par√°metro clave:

````python
def compare_lambda (decision):
    if decision == "mayorque":
        return lambda a, b: a if a > b else b
    elif decision == "menorque":
        return lambda a, b: b if b < a else a
    
compare_lambda("menorque")(3, 30)
````

````python
def compare_lambda (decision):
    if decision == "mayorque":
        return lambda a, b: max(a,  b)
    elif decision == "menorque":
        return lambda a, b: min(a, b)
    
compare_lambda("menorque")(3, 30)
````

## Funciones: funciones como retorno de otra funci√≥n
OTRO EJEMPLO en recursividad

````python
def greater_than (a, b):
    return max([a, b])

def less_than (a, b):
    return min([a, b])

def compare (decision):
    if decision == "mayorque":
        return greater_than
    elif decision == "menorque":
        return less_than
````

````python
compare("mayorque")(3, 30)

# 1. compare("mayorque") -> greater_than
# 2. greater_than(3, 30)
````

## Recursividad

Se denomina llamada recursiva (o recursi√≥n), a aquellas funciones que en su algoritmo, hacen referencia a s√≠ misma.

Las llamadas recursivas suelen ser muy √∫tiles en casos muy concretos, pero debido a su gran factibilidad de caer en iteraciones infinitas, se deben tomar las medidas preventivas adecuadas y solo utilizarlas cuando sea estrictamente necesario y no exista una forma alternativa viable de resolver el problema, evitando as√≠ la recursi√≥n.

Python soporta llamadas recursivas, permitiendo que una funci√≥n se llame a s√≠ misma, tal y como lo hace al llamar a otra funci√≥n.

````python
#PIDO LA FUNCION COMO DEVLUCI√ìN DE ESA MISMA FUNCI√ìN
def numero ():
    """Esta funci√≥n hace cosas"""
    print("Ejecutando funci√≥n")
    value = int(input (prompt = "Un n√∫mero mayor que 5: "))
    if (value > 5):
        return value
    return numero() #si no lo cumple entonces vuelve a ejecutar la funci√≥n

valor_mayor_que_cinco = numero()
valor_mayor_que_cinco
````
